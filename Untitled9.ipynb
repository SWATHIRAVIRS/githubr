```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

while True:

    text = input("Enter text (type 'exit' to stop): ")

    if text.lower() == "exit":
        break

    result = classifier(text)

    print("\nSentiment:", result[0]["label"])
    print("Confidence:", result[0]["score"])
    print("\n----------------------\n")
```
