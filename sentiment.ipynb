```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import LSTM,Embedding,Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import imdb
```

```python
vocab_size=10000
```

```python
(X_train,y_train),(X_test,y_test)=imdb.load_data(num_words=vocab_size)
print("Training samples:", len(X_train))
print("Example review (numbers):", X_train[0])
```

```python
max_len = 200
X_train = pad_sequences(X_train, maxlen=max_len)
X_test = pad_sequences(X_test, maxlen=max_len)
print("After padding shape:", X_train.shape)
```

```python
model=Sequential([
    Embedding(vocab_size,128,input_length=max_len),
    LSTM(128),
    Dense(1,activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,epochs=5,batch_size=64,validation_split=0.2)
```

```python
loss, acc = model.evaluate(X_test, y_test)
print("Test Accuracy:", acc)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.datasets import imdb

word_index = imdb.get_word_index()

sentence = "this movie is very good"

# convert words â†’ numbers
encoded = [word_index.get(word, 2) + 3 for word in sentence.lower().split()]

# padding
padded = pad_sequences([encoded], maxlen=200)

# prediction
pred = model.predict(padded)[0][0]

print("Score:", pred)

if pred > 0.5:
    print("Positive ğŸ˜Š")
else:
    print("Negative ğŸ˜¡")
```
