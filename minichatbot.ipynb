```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
tokenizer=AutoTokenizer.from_pretrained("microsoft/DialoGPT-small")
model=AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-small")
```

```python
chat_history_ids=None
#this stores past conversation as the conversation is yet started we set it none
while True:
    user_input=input("You:")
    new_input_ids=tokenizer.encode(user_input+tokenizer.eos_token,return_tensors="pt")
    bot_input_id=torch.cat([chat_history_ids,new_input_ids],dim=-1) if chat_history_ids is not None else new_input_ids

    #generate response
    chat_history_ids=model.generate(bot_input_id,max_length=100,pad_token_id=tokenizer.eos_token_id)
    #decode response
    response=tokenizer.decode(chat_history_ids[:,bot_input_id.shape[-1]:][0],skip_special_tokens=True)
    print("Bot:",response)
```
